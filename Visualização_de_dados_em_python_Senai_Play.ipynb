Aula 06 - Construção de gráfico histograma

```python
import matplotlib.pyplot as plt

quantidade = [40, 65, 30, 55]
meses = ['jan/2024', 'fev/2024', 'mar/2024', 'abr/2024']

fig, ax = plt.subplots(figsize=(10, 5));  # Descompactando a tupla retornada
ax.bar(meses, quantidade);  # (x, y)

ax.set_xlabel('Meses');  # Definindo o rótulo do eixo x
ax.set_ylabel('Valor Gasto');  # Definindo o rótulo do eixo y
ax.set_title('Valor historico dos gastos - 2024'); #atribuindo um titulo
ax.legend(title ='Gastos Combustivel',loc='lower right'); #Legenda


plt.show()
```

Aula 17 - Utilização de numpy e seaborn

```python
import numpy as np
import seaborn as sns
```

```python
#Iniciar um gerador de números pseudoaleatórios com uma semente fixa.
x= np.random.RandomState(1).normal(0,1, size=50)
```

```python
x
```

```python
ax = sns.kdeplot(x=x, fill=True, alpha=0.3, color='limegreen')
ax.axvline(x=1,ymin=0,ymax=1)
ax.axvline(x=-2,ymin=0,ymax=1)
```

```python
ax = sns.kdeplot(x=x, fill=True, alpha=0.3, color='limegreen')
ax.axvline(x=1,ymin=0.05,ymax=0.3)
```

```python
ax = sns.kdeplot(x=x, fill=True, alpha=0.3, color='limegreen')
ax.axhline(y=0.2,xmin=0.1,xmax=0.9)
```
